```python
import matplotlib  as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
%matplotlib inline
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
fig = plt.gcf()
print("Current figure: ", fig)
ax = plt.gca()
print("Current axes: ", ax)

ax.set_title("Cosine curve")
# Example: set a global figure title
fig.suptitle("Sine and Cosine", fontsize=14)
plt.show()
```
